```python
from sklearn.preprocessing import LabelEncoder # for 
```

```python
outlook	= LabelEncoder()
temp =LabelEncoder()
humidity=LabelEncoder()	
windy=LabelEncoder()
play=LabelEncoder()
```

```python
# fit transform converts categorial to number 
df['outlook']=outlook.fit_transform(df['outlook'])
df['temp']=outlook.fit_transform(df['temp'])
df['humidity']=outlook.fit_transform(df['humidity'])
df['windy']=outlook.fit_transform(df['windy'])
df['play']=outlook.fit_transform(df['play'])
```

```python
features_column = ["outlook","humidity","windy"]
X = df[features_column]
y = df.play 
# this are seprate out the dependent and independent variable 
```
